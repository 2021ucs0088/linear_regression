```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot
plt.plot(x, y)

# Labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Title
plt.title('Simple Line Graph')

# Show plot
plt.show()
```
