```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Salary_Data.csv')
```

```python
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
print(x.head())
```

```python
print(y.head())
```

```python
plt.scatter(x,y,color="red");
plt.title("sallery vs year of expeiencr");
plt.xlabel("year of exprinece");
plt.ylabel("sallery");
plt.show();
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.8);
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression();
```

```python
lr.fit(x_test,y_test)
```

```python
#model test below
```

```python
y_predict=lr.predict(x_test);
df=pd.DataFrame(y_test,y_predict);
```

```python
print(df)
```

```python
y_predict=lr.predict(x_train);
df=pd.DataFrame(y_test,y_predict);
```

```python
print(df)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
y_predict=lr.predict(x_test);
mse=mean_squared_error(y_test,y_predict)
print(mse)
mae=mean_absolute_error(y_test,y_predict);
print(mae)
r2=r2_score(y_test,y_predict);
print(r2);
```

# Visualize Training Set Results (with Regression Line)

```python
y_predict=lr.predict(x_train)
plt.scatter(x_train,y_train,color="red",label=" actual")
plt.scatter(x_train,y_predict,color="blue",label="predicted")
plt.plot(x_train,y_predict);
plt.legend()
plt.show()
```
